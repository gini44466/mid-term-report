```python
from transformers import pipeline

# 聊天用小型模型
chatbot = pipeline("text-generation", model="microsoft/DialoGPT-small")

# 翻譯模型
translator_zh_en = pipeline("translation", model="Helsinki-NLP/opus-mt-zh-en")
translator_en_zh = pipeline("translation", model="Helsinki-NLP/opus-mt-en-zh")
```

```python
# 聊天功能
def chat_tool(text):
    result = chatbot(text, max_length=200)
    return result[0]["generated_text"]

# 中文 ↔ 英文翻譯
def translate_text(text, lang):
    if lang == "英文":
        result = translator_zh_en(text)
        return result[0]['translation_text']
    elif lang == "中文":
        result = translator_en_zh(text)
        return result[0]['translation_text']
    else:
        return "目前只支援中文↔英文"
```

```python
import gradio as gr

with gr.Blocks(title="免金鑰 AI 百寶箱") as demo:
    gr.Markdown("## 🧰 AI 百寶箱（聊天 + 中文↔英文翻譯）")

    # 聊天分頁
    with gr.Tab("💬 聊天"):
        chat_in = gr.Textbox(label="輸入訊息")
        chat_out = gr.Textbox(label="AI 回覆")
        gr.Button("送出").click(chat_tool, chat_in, chat_out)

    # 翻譯分頁
    with gr.Tab("🌏 翻譯"):
        trans_in = gr.Textbox(label="輸入文字", lines=3)
        lang = gr.Dropdown(["英文", "中文"], label="選擇翻譯語言")
        trans_out = gr.Textbox(label="翻譯結果")
        gr.Button("開始翻譯").click(translate_text, [trans_in, lang], trans_out)

demo.launch()
```
